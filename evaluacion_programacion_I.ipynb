## **2. Agregar producto (4 pts)**

Crear la función `agregar_producto(inventario)` que:

1. Solicite al usuario el código del producto.

2. Use try–except para validar:
    - Que el código sea un entero.
    - Que no esté repetido (si está repetido → lanzar raise ValueError).

3. Solicite nombre, categoría, precio y cantidad.

4. Inserte el nuevo producto en el inventario.

> Debe usar al menos un método de listas o diccionarios.

```python
def agregar_producto(inventario):
    
    """
    """
    
    #  Solicitar código y validar que sea entero
    try:
        codigo = int(input("Ingresa el código del producto (debe ser un número entero): "))
    except ValueError:
        raise ValueError("Error: El código debe ser un número entero.")
        
    # Validación de código repetido
    for producto in inventario:
        if producto.get("codigo") == codigo:
            raise ValueError(f"Error: El código {codigo} ya existe en el inventario.")

    #Solicitar nombre y categoría
    nombre = input("Ingresa el nombre del producto: ")
    categoria = input("Ingresa la categoría: ")
    
    # Validación del precio y cantidad
    try:
        precio = float(input("Ingresa el precio: "))
        cantidad = int(input("Ingresa la cantidad en stock: "))
    except ValueError:
        raise ValueError("Error: El precio debe ser un número decimal y la cantidad un número entero.")
        
    # diccionario e insertar en la lista
    nuevo_producto = {
        "codigo": codigo,
        "nombre": nombre,
        "categoria": categoria,
        "precio": precio,
        "cantidad": cantidad
    }
    
    inventario.append(nuevo_producto)
    print(f"Producto '{nombre}' agregado con éxito.")
```

## **3. Buscar producto (4 pts)**

Crear la función `buscar_producto(inventario)` que:

1. Solicite un código.
2. Maneje los posibles errores:

    - Código no entero → capturar con except.
    - Producto no encontrado → lanzar raise KeyError.

3. Si lo encuentra, debe devolver los datos del producto.
4. Debe demostrar el uso de **indexación**, **in** o **keys()**, según la estructura elegida.

```python
def buscar_producto(inventario):
    # Solicitar el código y validar que sea entero con try-except
    try:
        codigo_buscado = int(input("Ingrese el código del producto a buscar: "))
    except :
        raise ValueError("Error: El código debe ser un número entero.")
        
    # no repetidpos 
    for producto in inventario:
        if producto.get("codigo") == codigo_buscado:
            return producto
            
    # devulve errror
    raise KeyError(f"Error: No se encontró el producto con código {codigo_buscado}")
```

## **4. Estadísticas del inventario (4 pts)**

Crear una función `estadisticas(inventario)` que calcule:

1. Cantidad total de productos distintos.
2. Número total de unidades en stock.
3. Precio promedio de los productos.
4. Una lista de comprensión con los productos cuyo stock sea mayor a 5.
5. Un conjunto (set) con todas las categorías sin repetir.

> Debe manejar excepciones internas si el inventario está vacío.

```python
def estadisticas(inventario):
    # Manejar excepción interna si el inventario está vacío
    if len(inventario) == 0:
        raise ValueError("Error: El inventario está vacío")
        
    total_productos = len(inventario)
    
    total_unidades = 0
    suma_precios = 0
    
    for producto in inventario:
        total_unidades += producto["cantidad"]
        suma_precios += producto["precio"]
        
    precio_promedio = suma_precios / total_productos
    
    # Lista de comprensión: productos mas de 5
    productos_stock_mayor_5 = [prod["nombre"] for prod in inventario if prod["cantidad"] > 5]
    
    # productos_stock_mayor_5 = []
    # for prod in inventario:
    #     if prod["cantidad"] > 5:
    #         productos_stock_mayor_5.append(prod["nombre"])

    
    # Conjunto 
    categorias = set()
    for prod in inventario:
        categorias.add(prod["categoria"])
        
    # Devolver un diccionario
    return {
        "total_productos": total_productos,
        "total_unidades": total_unidades,
        "precio_promedio": precio_promedio,
        "productos_stock_mayor_5": productos_stock_mayor_5,
        "categorias": categorias
    }
```

## **5. Exportar inventario (4 pts)**

Crear una función `exportar_inventario(inventario)` que:

1. Devuelva un diccionario donde:
    - Las claves sean los códigos de productos.
    - Los valores sean tuplas con (nombre, categoría, precio, cantidad).
2. Si existen datos inválidos (por ejemplo, una lista vacía o un producto mal formado), usar:
    - try–except
    - else
    - finally → imprimir un mensaje de cierre del proceso.

```python
def exportar_inventario(inventario):
    inventarioExportado = {}
    
    try:
        # Forzar un error si la lista está vacía
        if len(inventario) == 0:
            raise ValueError("El inventario está vacío.")
            
        for producto in inventario:
            # Clave: código. Valor: tupla con los demás datos
            codigo = producto["codigo"]
            tupla_datos = (producto["nombre"], producto["categoria"], producto["precio"], producto["cantidad"])
            
            inventarioExportado[codigo] = tupla_datos
            
    except ValueError:
        print("Ocurrió un problema con los datos al exportar")
       
        
    else:
        # Se ejecuta solo si el try fue exitoso
        print("Los datos han sido preparados para la exportación exitosamente.")
        return inventarioExportado
        
    finally:
        # Se ejecuta siempre al final
        print("s Finalizando el proceso de exportación ---")
```

## **Función principal `main`**

```python
def main():
    # Crear inventario inicial
    inventario = crear_inventario()
    print("Inventario inicial creado.")
    
    while True:
        print("\nOpciones:")
        print("1. Agregar producto")
        print("2. Buscar producto")
        print("3. Ver estadísticas")
        print("4. Exportar inventario")
        print("5. Salir")
        
        opcion = input("Seleccione una opción: ")
        
        match opcion:
            case '1':
                agregar_producto(inventario)
            case '2':
                try:
                    producto = buscar_producto(inventario)
                    if producto:
                        print(f"Producto encontrado: {producto}")
                except KeyError as e:
                    print(e)
            case '3':
                try:
                    stats = estadisticas(inventario)
                    print(f"Total productos: {stats['total_productos']}")
                    print(f"Total unidades: {stats['total_unidades']}")
                    print(f"Precio promedio: {stats['precio_promedio']:.2f}")
                    print(f"Productos con stock > 5: {stats['productos_stock_mayor_5']}")
                    print(f"Categorías: {stats['categorias']}")
                except ValueError as e:
                    print(e)
            case '4':
                exportado = exportar_inventario(inventario)
                if exportado:
                    print("Inventario exportado:", exportado)
            case '5':
                break
            case _:
                print("Opción inválida.")

if __name__ == "__main__":
    main()
```
